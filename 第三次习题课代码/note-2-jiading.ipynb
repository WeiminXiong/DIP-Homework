# Python频域滤波

```python
#理想低通滤波器

#计算二维点pa和pb之间的距离
def cal_distance(pa,pb):
    from math import sqrt
    dis=sqrt((pa[0]-pb[0])**2+(pa[1]-pb[1])**2)
    return dis

def make_transform_matrix(d,image):
    transform_matrix=np.zeros(image.shape)
    center_point=tuple(map(lambda x:(x-1)/2,image.shape))
    for i in range(transform_matrix.shape[0]):
        for j in range(transform_matrix.shape[1]):
            dis=cal_distance(center_point,(i,j))
            if dis<=d:
                transform_matrix[i,j]=1
            else:
                transform_matrix[i,j]=0
    return transform_matrix
```

```python
#d=10.0
transform_matrix=make_transform_matrix(d=10.0,image=lena)
plt.imshow(np.abs(transform_matrix),cmap='gray')
plt.show()
```

```python
#获得lena的频域中
fft_lena=np.fft.fft2(lena)
#shift
fft_lena_shift=np.fft.fftshift(fft_lena)
#norm
fft_lena_shift_norm=np.log(np.abs(fft_lena_shift)+1)
#plt
plt.imshow(fft_lena_shift_norm,cmap='gray')
plt.show()
```

```python
#将transform矩阵apply到lena的频域中
fft_lena_shift_norm_transformed=fft_lena_shift*transform_matrix
#反变换
f_ishift=np.fft.ifftshift(fft_lena_shift_norm_transformed)
img_back=np.fft.ifft2(f_ishift)
#img_back=cv2.magnitude(img_back[:,:,0],img_back[:,:,1])
plt.imshow(np.abs(img_back),cmap='gray')
plt.show()
```

```python
## 高斯低通滤波器
def make_transform_matrix(d,image):
    transform_matrix=np.zeros(image.shape)
    center_point=tuple(map(lambda x:(x-1)/2,image.shape))
    for i in range(transform_matrix.shape[0]):
        for j in range(transform_matrix.shape[1]):
            dis=cal_distance(center_point,(i,j))
            transform_matrix[i,j]=np.exp(-(dis**2)/(2*(d**2)))
    return transform_matrix
```

高斯模糊:https://baike.baidu.com/item/%E9%AB%98%E6%96%AF%E6%A8%A1%E7%B3%8A/10885810

>高斯模糊（英语：Gaussian Blur），也叫高斯平滑，是在Adobe Photoshop、GIMP以及Paint.NET等图像处理软件中广泛使用的处理效果，通常用它来减少图像噪声以及降低细节层次。这种模糊技术生成的图像，其视觉效果就像是经过一个毛玻璃在观察图像，这与镜头焦外成像效果散景以及普通照明阴影中的效果都明显不同。高斯平滑也用于计算机视觉算法中的预先处理阶段，以增强图像在不同比例大小下的图像效果（参见尺度空间表示以及尺度空间实现）。 从数学的角度来看，图像的高斯模糊过程就是图像与正态分布做卷积。由于正态分布又叫作高斯分布，所以这项技术就叫作高斯模糊。图像与圆形方框模糊做卷积将会生成更加精确的焦外成像效果。由于高斯函数的傅立叶变换是另外一个高斯函数，所以高斯模糊对于图像来说就是一个低通滤波器。

```python
# 高斯高通滤波器

def make_transform_matrix(d,image):
    transform_matrix=np.zeros(image.shape)
    center_point=tuple(map(lambda x:(x-1)/2,image.shape))
    for i in range(transform_matrix.shape[0]):
        for j in range(transform_matrix.shape[1]):
            dis=cal_distance(center_point,(i,j))
            transform_matrix[i,j]=1-np.exp(-(dis**2)/(2*(d**2)))
    return transform_matrix
```

```python
#陷波滤波器
#https://github.com/imdeep2905/Notch-Filter-for-Image-Processing
#理想陷波滤波器   
import random
def apply_filter(image, num_of_points, d0):
    
    f=cv2.dft(np.float32(image),flags=cv2.DFT_COMPLEX_OUTPUT)
    fshift=np.fft.fftshift(f)
    
    points=np.zeros((num_of_points,2))
    h,w,_=fshift.shape
    for i in range(num_of_points):
        points[i][0]=random.randint(0,h)
        points[i][1]=random.randint(0,w)
    
    plt.subplot(1,3,1)
    plt.imshow(image,cmap='gray')
    plt.title('original')
    
    m = fshift.shape[0]
    n = fshift.shape[1]
    for u in range(m):
        for v in range(n):
            for d in range(len(points)):
                u0 = points[d][0]
                v0 = points[d][1]
                u0, v0 = v0, u0
                d1 = pow(pow(u - u0, 2) + pow(v - v0, 2), 1)
                d2 = pow(pow(u + u0, 2) + pow(v + v0, 2), 1)
                if d1 <= d0 or d2 <= d0:
                    fshift[u][v] *= 0.0
    magnitude_spactrum=20*np.log(cv2.magnitude(fshift[:,:,0],fshift[:,:,1]))
    f_ishift = np.fft.ifftshift(fshift)
    img_back=cv2.idft(f_ishift)
    img_back=cv2.magnitude(img_back[:,:,0],img_back[:,:,1])
    
    plt.subplot(1,3,2)
    plt.imshow(abs(magnitude_spactrum),cmap='gray')
    plt.title('after filter')
    
    plt.subplot(1,3,3)
    plt.imshow(img_back,cmap='gray')
    plt.title('after filter-idft')    
    plt.show()

num_of_points=8
r=150

apply_filter(lena,num_of_points,r)
```
